```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot (diagramas de dispersion)
from msilib.schema import Font


x = np.arange(1,20)
y = x**2
y1 = x**2 +5*x

plt.scatter(x,y,s=10, color='red',marker='s',label='x**2')
plt.scatter(x,y1,s=10, color='blue',marker='s',label='x**2')
plt.legend(title='Funciones') #para titulo
plt. title ('TIPOS DE FUNCIONES',Fontsize=20) #Fontsize para resaltar el titulo
plt.xlabel ('x',color='green')
plt.ylabel ('f(x)')
#plt.xlim(0,5) para modificar el eje
plt.grid()

plt.text(5,300,'este texto',color='brown') #plt.text para incluir texto

```
